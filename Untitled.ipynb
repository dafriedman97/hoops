```python
import numpy as np
import pandas as pd
import os
```

```python
pbp_dir = "data/play_by_play/2021-22"
pbps = [pd.read_csv(pbp_dir + "/" + file) for file in os.listdir(pbp_dir) if file.endswith(".csv")]
pbps = pd.concat(pbps).reset_index(drop=True)
```

```python
home, vis = "CHI", "MIL"
pbps_teams = pbps.loc[(pbps['home'].isin([home, vis])) | (pbps['vis'].isin([home, vis]))].reset_index(drop=True)
scores = pbps_teams.loc[pbps_teams['event'].isin(['fg_make', 'fg3_make', 'ft_make'])].fillna("")
first_scores = scores.groupby('game_id').first()
first_scores
```
